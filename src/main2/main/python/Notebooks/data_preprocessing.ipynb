```python
# Pandas for managing datasets
import pandas as pd
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
import numpy as np

# Read dataset
df = pd.read_csv('C:\\Ashish\\Clustering-Bin-Packing\\src\\main\\resources\\outbound_final.csv', index_col=0)

# Display first 5 observations
df.head()
```

```python
data_x = []
data_y = []
for i in df['collection_geoloc']:
    data_temp = i.strip().split(',')
    data_x.append(data_temp[0])
    data_y.append(data_temp[1])
    
data_x_float = [float(i) for i in data_x]
data_y_float = [float(i) for i in data_y]
x_array = np.asarray(data_x_float)
y_array = np.asarray(data_y_float)
```

```python
data_x1 = []
data_y1 = []
for i in df['delivery_geoloc']:
    data_temp1 = i.strip().split(',')
    data_x1.append(data_temp1[0])
    data_y1.append(data_temp1[1])
    
data_x_float1 = [float(i) for i in data_x1]
data_y_float1 = [float(i) for i in data_y1]
x_array1 = np.asarray(data_x_float1)
y_array1 = np.asarray(data_y_float1)
```
